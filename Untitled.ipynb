```python
import matplotlib.pyplot as plt, pandas as pd
import numpy as np, os, sys, networkx as nx, warnings
import mercator
from itertools import product
from tqdm import tqdm
from pathlib import Path

# fn = "./test_watts.inf_coord"
# with open(fn) as f:
#     for i in range(8):
#         sline=next(f).strip().split()
#     beta=float(next(f).strip().split()[3])
#     mu=float(next(f).strip().split()[3])
#     R=float(next(f).strip().split()[3])

def hyp_dist(x, y, mu, R):
    da = np.pi - np.fabs( np.pi - np.fabs( x.theta.iloc[0] - y.theta.iloc[0] ) )
    return (R * da) / (mu * x.kappa.iloc[0] * y.kappa)

def jaccard(a, b):
    return len(a.intersection(b)) / (len(a.union(b)))

def ang_diff(x, y, c = 2 * np.pi):
    return c - np.fabs(np.mod(np.fabs(x.theta.iloc[0] - y.theta.iloc[0]), 2*c) - c)

def ang(x, y):
    delta = float(x.theta.iloc[0]) - float(y.theta.iloc[0])
    # delta /= np.pi * 2
    # delta = delta * 2 - 1
    return np.fabs(delta) / (2 * np.pi)



def load_inf(fn):
    fn = Path(fn.resolve()).with_suffix(".inf_coord")
    # fn = "./" + str(fn.resolve()).split(".")[0] + ".inf_coord"
    print(fn)
    data = {}
    for idx, label in enumerate("name kappa theta rad".split()):
        data[label] = np.loadtxt(fn, usecols=[idx], dtype = object)
    data = pd.DataFrame(data)
    data.convert_dtypes()
    return data
def h_dist(xx, xy, yx, yy):
    return np.arccosh(np.cosh(xy)*np.cosh(yx - xx) * np.cosh(yy) - np.sinh(xy)*np.sinh(yy))
def load_edges(g, data):
    tmp = []
    c = nx.edge_betweenness_centrality(g, k=None, normalized=True, weight=None, seed=None)
    for k, v in g.edges():
        x, y = data[data.name == k], data[data.name == v]
        rx, ry = float(x.rad.iloc[0]), float(y.rad.iloc[0])
        th1 = float(x.theta.iloc[0])
        th2 = float(y.theta.iloc[0])
        xx = rx * np.cos(th1)
        xy = rx * np.sin(th1)
        
        yx = ry * np.cos(th2)
        yy = ry * np.sin(th2)
        
        rad = h_dist(xx, xy, yx, yy)
        d = ang(x, y)
        a, b = set(g.neighbors(k)), set(g.neighbors(v))
        j = jaccard(a, b)
        h = dict(a = k, b = v, angular = d, rad = rad,
                 jaccard = j, betweenness = c[(k, v)])
        tmp.append(h)
    return pd.DataFrame(tmp)

def read_parameters(fn):
    params = {}
    for c in fn.split("_")[1:]:
        label, thing = c.split("=")
        params[label] = float(thing)
    return params
    
def load_graphs(path: str):
    p = Path(path)
    print(p.resolve())
    data = {}
    for fn in tqdm(p.iterdir()):
        if fn.suffix == ".edge":
            # print(f"Reading {fn}\r")
            g = nx.read_edgelist(fn.resolve())
            # print(len(g))
            d = load_inf(fn.resolve())
            edge = load_edges(g, d)
            # edge = []
            # print(fn.stem)
            params = read_parameters(fn.stem)
            data[str(fn.stem)] = dict(g = g, data = d,
                                      edge = edge, **params)
    print("loaded data")
    return data
graphs = load_graphs("./")
```

```python
# retrieve unique conditions
def get_unique_conditions(graphs):
    us = dict(p = set(), trial = set(), k = set())
    for k, v in graphs.items():
        for idx, label in enumerate("p trial k".split()):
            us[label].add(v[label])
    return us
uparams = get_unique_conditions(graphs)
```
